```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()
tf.compat.v1.experimental.output_all_intermediates(True)
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
#sess = tf.compat.v1.Session(config=config)
sess = tf.compat.v1.InteractiveSession(config=config)

import keras.backend.tensorflow_backend as tfback
def _get_available_gpus():
    """Get a list of available gpu devices (formatted as strings)

    # Returns
        A list of available GPU devices.
    """
    #global _LOCAL_DEVICES
    if tfback._LOCAL_DEVICES is None:
        devices = tf.config.list_logical_devices()
        tfback._LOCAL_DEVICES = [x.name for x in devices]
    return [x for x in tfback._LOCAL_DEVICES if 'device:gpu' in x.lower()]
tfback._get_available_gpus = _get_available_gpus
tfback._get_available_gpus()
tf.config.list_logical_devices()

import keras
#keras.backend.clear_session()

from copy import deepcopy
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PCONV_UNET import *
from augmentation.geometric_distortion import RandomTranslation, RandomScalingAndRotation, RandomShearing
from augmentation.augmentation import SequentialTransform
```

```python
model = PConvUnet(256 , 512 , vgg_weights='pytorch_to_keras_vgg16.h5')
model.summary()
#model.load('gan9.h5', train_bn=False)
```

```python
def augmentation(masked , mask , ori):
    h, w = masked.shape[:2]
    
    translation = RandomTranslation((-0.1, 0.1), (-0.1, 0.1))
    rotation_and_scaling = RandomScalingAndRotation((-10, 10), (0.9, 1.1))
    shearing = RandomShearing((-0.1, 0.1), (-0.1, 0.1))
    sequential = SequentialTransform([translation, rotation_and_scaling, shearing],(w, h))

    aug_masked , aug_mask , aug_ori = sequential.apply_transform(masked , mask , ori , border_mode=cv2.BORDER_CONSTANT)
    return aug_masked , aug_mask , aug_ori
```

```python
def plot_sample_data(masked, mask, ori, middle_title='Raw Mask'):
    _, axes = plt.subplots(1, 3, figsize=(20, 5))
    axes[0].imshow(np.array(masked[:,:,:]*255 , np.uint8))
    axes[0].set_title('Masked Input')
    axes[1].imshow(np.array(mask[:,:,:]*255 , np.uint8))
    axes[1].set_title(middle_title)
    axes[2].imshow(np.array(ori[:,:,:]*255 , np.uint8))
    axes[2].set_title('Target Output')
    plt.show()
```

```python
def get_data(x1_paths , x2_paths , y_paths , i , aug = False):
    y = '../data/Dataset/clean/' + y_paths[i]
    y = cv2.imread(y)
    y = cv2.cvtColor(y, cv2.COLOR_BGR2RGB)
    y = cv2.resize(y , (512 , 256))
    img = '../data/Dataset/damaged/' + x1_paths[i]
    img = cv2.imread(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img , (512 , 256))
    mask = '../data/Dataset/masks/' + x2_paths[i]
    mask = cv2.imread(mask)
    mask = cv2.bitwise_not(mask)
    mask = cv2.resize(mask , (512 , 256))
    masked_img = deepcopy(img)
    masked_img[mask==0] = 255#plot_sample_data(masked_img, mask, y, middle_title='Raw Mask')
    if aug :
        masked_img , mask , y = augmentation(masked_img , mask , y)
    return [np.expand_dims(masked_img , 0)/255.0 , np.expand_dims(mask , 0)/255.0] , np.expand_dims(y , 0)/255.0
```

```python
def get_mask_spots(path , lower_range , higher_range):
    img = cv2.imread(path)
    img = cv2.resize(img , (768,512))
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, lower_range, higher_range)
    mask  = cv2.morphologyEx(cv2.medianBlur(mask , 3) , cv2.MORPH_CLOSE, np.ones((3,3) , np.uint8))
    return mask

def test():
    damaged_paths = os.listdir('../Damaged/')
    damaged_path = damaged_paths[np.random.randint(len(damaged_paths))]
    values = np.split(np.array(damaged_path[:-4].split('_'),np.int32) , 2)
    mask = get_mask_spots('../Damaged/'+ damaged_path , values[0] , values[1])
    img  = cv2.resize(cv2.cvtColor(cv2.imread('../Damaged/' + damaged_path), cv2.COLOR_BGR2RGB), (512 , 256))
    mask = cv2.resize(cv2.cvtColor(mask , cv2.COLOR_GRAY2BGR) , (512 , 256))
    mask = cv2.bitwise_not(mask)
    return [np.expand_dims(img , 0)/255.0 , np.expand_dims(mask , 0)/255.0 ]
x = test()
plot_sample_datargb(x[0][0], model.predict(x)[0] ,x[1][0],middle_title='Prediction')
```
